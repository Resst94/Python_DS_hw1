```python
## Завдання №16: Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 
## та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.

matrix = np.random.random((3, 3))
row_sums = np.sum(matrix, axis=1, keepdims=True)

print("Матриця:\n", matrix)
print("Вектор-стовпчик, що містить суму елементів кожного рядка матриці:\n", row_sums)
```

```python
## Завдання №17: Створіть матрицю розміром 3x4 з довільними цілими числами і створіть матрицю з квадратами цих 
## чисел.

matrix = np.random.randint(-100, 100, (3, 4))

print("Матриця:\n", matrix)
print("Матриця з квадратами чисел v1:\n", np.square(matrix))
print("Матриця з квадратами чисел v2:\n", matrix ** 2)
print("Матриця з квадратами чисел v3:\n", matrix * matrix)
```
